```python
comparison = dec_2016.join(nov_2017, lsuffix='_2016', rsuffix='_2017')
```

```python
print(comparison)
```

```python
dec_2016.columns = [('2016_mean')]
nov_2017.columns = [('2017_mean')]
```

```python
comparison = dec_2016.join(nov_2017)
comparison['change_%'] = ((comparison['2017_mean'] - comparison['2016_mean']) / comparison['2016_mean'] * 100).map(lambda x: f"{x:+.1f}%")
```
